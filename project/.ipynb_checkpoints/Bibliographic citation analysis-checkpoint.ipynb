```python
import requests
from bs4 import BeautifulSoup

# Set the search URL
search_term = "heart disease"
SEARCH_URL = f"https://pubmed.ncbi.nlm.nih.gov/?term={search_term}&filter=simsearch2.ffrft&size=200"

# Get the search results page
response = requests.get(SEARCH_URL)
soup = BeautifulSoup(response.content, 'html.parser')

# Extract links for each article from the search results page
article_links = []
for link in soup.find_all('a', class_='docsum-title', href=True):
    article_links.append(link['href'])

# Walk through the links and extract the details
articles_data = []
def safe_extract_text(tag):
    return tag.text.strip() if tag else None
def safe_extract_text2(tag):
    return tag.next_sibling.strip() if tag else None
for link in article_links:
    article_url = f"https://pubmed.ncbi.nlm.nih.gov{link}"
    response = requests.get(article_url)
    article_soup = BeautifulSoup(response.content, 'html.parser')
    
    
    # Extract author name, date, keywords, literature citation information, etc
    article_title = [safe_extract_text(article_soup.find('h1', class_='heading-title'))]
    authors = [safe_extract_text(article_soup.find('a', class_='full-name'))]
    dates = [safe_extract_text(article_soup.find('span', class_='cit'))]
    keywords = [safe_extract_text2(article_soup.find('strong', class_='sub-title'))]
    references = [safe_extract_text(reference) for reference in article_soup.find_all('li', class_='skip-numbering')][:8]
    
    articles_data.append({
        'title': article_title,
        'authors': authors,
        'date': dates,
        'keywords': keywords,
        'references': references,
        'url': article_url
    })

# Print the first eight extracted data
for article in articles_data[:8]:
    print(article)
```

```python
df_keywords = pd.DataFrame(dict_list_modified)

df_keywords['year'] = df_keywords['date'].str.extract(r'(\d{4})')

df_keyword_counts = df_keywords.explode('keywords').groupby(['year', 'keywords']).size().unstack(fill_value=0)

plt.figure(figsize=(14, 8))
df_keyword_counts.plot(kind='bar', stacked=True, figsize=(14, 7), colormap="tab20", ax=plt.gca())

plt.title("Research Trends Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Articles")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.xticks(list(range(0, len(df_keyword_counts), 5)), df_keyword_counts.index[::5], rotation=45)
plt.show()
#The chart shows the research trends from year to year
```
